# Procedure for the trees

~Height(tree)

if tree=nil:
    return 0
return 1+Max(Height(tree.left),Height(tree.right))

~Size(tree)

if tree=nil:
    return 0
return 1+ Size(tree.left)+ Size(tree.right)


#  Traversing The  Trees

There are two methods to walk on these trees

1 Depth first method : We completely traverse one sub-tree before exploring a sibling sub-tree
2 Breadth first method : We traverse all the nodes at one level before progressing to the next level
