# Times for common operation

O[n] means we need to make some number of operations to execute this particular task
O[1] means that we just need only one operation to finish the task in hand

In an array the alteration in the array elements can be calculated by the O[ ] operations so 

If we want to add a element or remove a element from begining the we will need O[n] operations , n here states the number of operations required to attain the final goal thus we need to stick to a fact that we cant leave holes in between the array so we need to shift the element accordingly to achieve our aim.

Same goes for the updating element in the middle it takes O[n] operations

The fastest updation process in the array is adding or removing element from the End it only takes O[1] operations.

So we can conclude that Arrays are great if we need to update any positions in the end but its expensive if we need to change any value in the middle or the begining

The greatest advantage to the array is its constant access time advantage

constant time to add/ remove at the end 
Linear time to add and remove at an arbitrary location


#  Singly Linked List


links in a chain , it has a head to a node and update link to the next node key

head is pointer that connects to a node and the node has two parts one is the value and the another is the pointer which points to the next element untill the nil appears.

[*] Operations provided by the linked list api & [*] Times of the Basic Operations

1 PushFront (Key)   -  Add to the front - O(1)
2 Key TopFront ()   -  Return Front Item 
3 PopFront()        -  Remove front item  - O(1)
4 PushBack(Key)     -  Add to the back (It is also known as Append)  - O(n) [if no tail given] else O(1)
5 Key TopBack()     -  Return Back Item
6 PopBack()         -  Remove Back item - O(n) [even if the tail is given cause we dont have pointer back from the last elem to the second last so we need to go ele by ele thus n operations ]
7 Boolean Find(Key) -  Is the element/ Key in the list
8 Erase(Key)        -  remove key from list 
9 Boolean Empty()   -  Check that list is empty or not
10 AddBefore(Node,Key) - Add key before the node
11 AddAfter(Node,Key) - Add key after the node
